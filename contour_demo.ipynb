```python
import skimage
import cv2
import numpy as np
import matplotlib.pyplot as plt
mask_path='/data/fandongxuan/temp/dataset/images/output/mask/mask_00000027.png'
# display mask
mask = skimage.io.imread(mask_path)
mask[mask != 255] = 0
# replace 0 values with 255 and 255 with 0
mask = 255 - mask
img,_=cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_TC89_KCOS)
# display contours
plt.imshow(mask, cmap='gray')
for i in range(len(img)):
    plt.scatter(img[i][:,0,0], img[i][:,0,1], c='r', s=1)
# adjust axis to zoom in
# plt.axis([275, 350, 300, 200])
plt.show()
```
